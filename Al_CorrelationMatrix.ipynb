```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

corr_down = pd.read_csv("correlation_matrix_Al_down.csv")

par_names_down = corr_down.columns.tolist()
corr_down.index = par_names_down

# Heatmap
plt.figure(figsize=(7, 6))
sns.heatmap(corr_down, annot=True, cmap='coolwarm', vmin=-1, vmax=1, square=True, linewidths=0.5)
plt.title("Matrice di Correlazione del Fit (Al down)")
plt.tight_layout()
plt.show()

```

```python
corr_up = pd.read_csv("correlation_matrix_Al_up.csv")

par_names_up = corr_up.columns.tolist()
corr_up.index = par_names_up

# Heatmap
plt.figure(figsize=(7, 6))
sns.heatmap(corr_up, annot=True, cmap='coolwarm', vmin=-1, vmax=1, square=True, linewidths=0.5)
plt.title("Matrice di Correlazione del Fit (Al up)")
plt.tight_layout()
plt.show()
```
